```python
fs.dtypes
```

```python
fs[fs['Location_State'] == ' CA']['Salary Estimate'].iloc[3].split()[0].split('-')[0][1:-1]
```

```python
fs['low pay'] = fs['Salary Estimate'].apply(lambda x: x.split()[0].split('-')[0][1:-1])
```

```python
fs['high pay'] = fs['Salary Estimate'].apply(lambda x: x.split()[0].split('-')[1][1:-1])
```

```python
fs.head()
```

```python
fs = fs.astype({"low pay": int, "high pay": int})
```

```python
fs['Average Pay'] = (fs['low pay'] + fs['high pay'])/2
```

TIME TO CHECK FIVE STAR COMPANIES WITH THE HIGHEST AND LOWEST PAY 

```python
fs.nlargest(5,'Average Pay')
```

```python
fs.nsmallest(5,'Average Pay')
```
